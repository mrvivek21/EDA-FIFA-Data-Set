```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fifa=pd.read_csv('fifa data.csv')
fifa
```

```python
for col in fifa.columns: 
    print(col) 
```

```python
fifa['nationality'].value_counts()
```

```python
fifa['nationality'].value_counts()[0:10]
```

```python
fifa['nationality'].value_counts()[0:5].keys()
```

```python
plt.figure(figsize=(7,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color="b")
plt.xlabel('Nationality',fontweight='bold')
plt.ylabel('Players',fontweight='bold')
plt.title('No of Players as per Nationality',fontweight='bold')
plt.show()  
```

```python
top_5_nationalities = fifa['nationality'].value_counts().head(5)
```

```python
plt.pie(top_5_nationalities, labels=top_5_nationalities.index, autopct='%0.1f%%')
plt.title('Top 5 Players Nationalities Distribution', fontweight='bold')
plt.show()
```

```python
player_salary = fifa[['short_name','wage_eur']]
player_salary
```

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head()
```

```python
plt.figure(figsize=(9,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=["blue","green","red","pink","orange"])
plt.xlabel('Players Name',fontweight='bold')
plt.ylabel('Salarys',fontweight='bold')
plt.title('Players Salaries in EUR',fontweight='bold')
plt.show()
```

```python
player_shooting= fifa[['short_name','shooting']]
player_shooting
```

```python
player_shooting.sort_values(by=['shooting'],ascending=False).head()
```

```python
sort = player_shooting.sort_values(by='shooting', ascending=False)
top = sort.head(10)
top
```

```python
plt.figure(figsize=(9, 5))
plt.bar(top['short_name'], top['shooting'],color="r")
plt.xlabel('Player\'s Name', fontweight='bold')
plt.ylabel('Shooting %', fontweight='bold')
plt.title('Top 10 Player Shooting', fontweight='bold')
plt.xticks(rotation=45)
plt.show()
```

```python
England=fifa[fifa['nationality']=='England']
England.head(10)
```

```python
England.sort_values(by=['wage_eur'],ascending=False).head()
```

```python
England[['short_name','wage_eur']].sort_values(by=['wage_eur'],ascending=False).head()
```

```python
sorted_df = England.sort_values(by='wage_eur', ascending=False)
```

```python
top_5 = sorted_df.head(5)
top_5
```

```python
plt.figure(figsize=(9, 5))
plt.bar(top_5['short_name'], top_5['wage_eur'], color=["red", "blue", "green", "pink", "orange"])
plt.xlabel('Player\'s Name', fontweight='bold')
plt.ylabel('Salary (EUR)', fontweight='bold')
plt.title('Top 5 Player Salaries in EUR', fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(8, 7))
plt.pie(top_5['wage_eur'], labels=top_5['short_name'], autopct='%0.1f%%')  
plt.title('Players Salaries', fontweight='bold')
plt.show()
```
